```python
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
def data_gen(img_folder1,img_folder2, mask_folder, batch_size, resize_size):
  c = 0
  n = os.listdir(img_folder1)

  while (True):
    img = np.zeros((batch_size, resize_size, resize_size, 6)).astype('float')
    mask = np.zeros((batch_size, resize_size, resize_size,2)).astype('float')

    for i in range(c, c+batch_size): #initially from 0 to 16, c = 0.

      train_img1 = cv2.imread(img_folder1+'/'+n[i])
      train_img1 = cv2.cvtColor(train_img1, cv2.COLOR_BGR2RGB)/255.
      train_img1 = cv2.resize(train_img1, (resize_size, resize_size))# Read an image from folder and resize

      train_img2 = cv2.imread(img_folder2+'/'+n[i])
      train_img2 = cv2.cvtColor(train_img2, cv2.COLOR_BGR2RGB)/255.
      train_img2 = cv2.resize(train_img2, (resize_size, resize_size))# Read an image from folder and resize

      train_img=np.dstack((train_img1,train_img2))

      img[i-c] = train_img #add to array - img[0], img[1], and so on.



      train_mask = cv2.imread(mask_folder+'/'+n[i], cv2.IMREAD_GRAYSCALE)/255.
      train_mask = cv2.resize(train_mask, (resize_size, resize_size))
      train_mask = train_mask.reshape(resize_size, resize_size) # Add extra dimension for parity with train_img size [512 * 512 * 3]
      #train_mask = to_categorical(train_mask, num_classes=2, dtype="float32")
      train_mask2=1-train_mask
      train_mask=np.stack((train_mask2, train_mask),axis=-1)

    #   print(train_mask.shape)
      mask[i-c] = train_mask

    c+=batch_size
    if(c+batch_size>=len(os.listdir(img_folder1))):
      c=0
    #   random.shuffle(n)  # print "randomizing again"
    yield img, mask

print("Train Data loaded")
```

```python

def conv3x3(x, out_filters, strides=(1, 1)):
    x = Conv2D(out_filters, 3, padding='same', strides=strides, use_bias=False, kernel_initializer='he_normal')(x)
    return x


def Conv2d_BN(x, nb_filter, kernel_size, strides=(1, 1), padding='same', use_activation=True):
    x = Conv2D(nb_filter, kernel_size, padding=padding, strides=strides, kernel_initializer='he_normal')(x)
    x = BatchNormalization(axis=3)(x)
    if use_activation:
        x = Activation('relu')(x)
        return x
    else:
        return x

#Left
classes=2
channel=6
r=128
input=Input(shape=(r, r, channel))
input1=input[:,:,:,0:3]
input2=input[:,:,:,3:6]

conv1_1_L = Conv2D(64, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(input1)
conv1_1_L = BatchNormalization(axis=3)(conv1_1_L)
conv1_1_L = Activation('relu')(conv1_1_L)
conv1_1_L = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv1_1_L)
#drop1_2_L = Dropout(0.25)(conv1_1_L)

# conv2_x  input size:64
conv2_1_L = Conv2D(128, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv1_1_L)
conv2_1_L = BatchNormalization(axis=3)(conv2_1_L)
conv2_1_L = Activation('relu')(conv2_1_L)
conv2_1_L = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv2_1_L)
#drop2_2_L = Dropout(0.5)(conv2_1_L)

# conv3_x  input size:32
conv3_1_L = Conv2D(256, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv2_1_L)
conv3_1_L = BatchNormalization(axis=3)(conv3_1_L)
conv3_1_L = Activation('relu')(conv3_1_L)
conv3_1_L = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv3_1_L)
#drop3_2_L = Dropout(0.5)(conv3_1_L)
# conv3_x  input size:16
conv4_1_L = Conv2D(512, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv3_1_L)
conv4_1_L = BatchNormalization(axis=3)(conv4_1_L)
conv4_1_L = Activation('relu')(conv4_1_L)
conv4_1_L = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv4_1_L)
drop4_2_L = Dropout(0.5)(conv4_1_L)

#spatial attention left
c1=32
c1=int(drop4_2_L.shape[3]/8)
c=4
c=drop4_2_L.shape[2]
d=256
d=drop4_2_L.shape[3]
fa=Conv2D(c1,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_L)
fb=Conv2D(c1,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_L)
fc=Conv2D(d,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_L)
fa1=tf.reshape(fa,[-1,c*c,c1])
fb1=tf.reshape(fb,[-1,c*c,c1])
fb1=tf.transpose(fb1, perm=[0, 2, 1])
fs=tf.linalg.matmul(fa1,fb1)
fs1=tf.nn.softmax(fs)
fc1=tf.reshape(fc,[-1,c*c,d])
t=tf.linalg.matmul(fs1,fc1)
t=tf.reshape(t,[-1,c,c,d])
#Fsa_1=0.01*t+drop4_2_L
#define eta left
e_l = tf.Variable(0.)
Fsa_1=e_l*t+drop4_2_L

#channel attention Left

fch=tf.reshape(drop4_2_L,[-1,c*c,d])
fch1=tf.transpose(fch, perm=[0, 2, 1])
F=tf.linalg.matmul(fch,fch1)
F=tf.nn.softmax(F)
s=tf.linalg.matmul(fch1,F)
s=tf.reshape(s,[-1,c,c,d])
#Fsc_1=0.01*s+drop4_2_L

#define gama left
g_l = tf.Variable(0.)
Fsc_1=g_l*s+drop4_2_L

#channel and spatial attention fusion in left in parallel form
Z_Left=Fsc_1+drop4_2_L
Z_Left=Conv2D(512,3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(Z_Left)
ZZ=Fsa_1+Z_Left
ZZ=Conv2D(512,3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(ZZ)
R_Left=drop4_2_L+ZZ

Fs_merge_1 = R_Left

# Right
conv1_1_R = Conv2D(64, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(input2)
conv1_1_R = BatchNormalization(axis=3)(conv1_1_R)
conv1_1_R = Activation('relu')(conv1_1_R)
conv1_1_R = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv1_1_R)
#drop1_2_R = Dropout(0.25)(conv1_1_R)

# conv2_x  input size:64
conv2_1_R = Conv2D(128, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv1_1_R)
conv2_1_R = BatchNormalization(axis=3)(conv2_1_R)
conv2_1_R = Activation('relu')(conv2_1_R)
conv2_1_R = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv2_1_R)
#drop2_2_R = Dropout(0.5)(conv2_1_R)

# conv3_x  input size:32
conv3_1_R = Conv2D(256, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv2_1_R)
conv3_1_R = BatchNormalization(axis=3)(conv3_1_R)
conv3_1_R = Activation('relu')(conv3_1_R)
conv3_1_R = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv3_1_R)
#drop3_2_R = Dropout(0.5)(conv3_1_R)
# conv3_x  input size:16
conv4_1_R = Conv2D(512, 5, strides=(1, 1), padding='same', use_bias=False, kernel_initializer='he_normal')(conv3_1_R)
conv4_1_R = BatchNormalization(axis=3)(conv4_1_R)
conv4_1_R = Activation('relu')(conv4_1_R)
conv4_1_R = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same')(conv4_1_R)
drop4_2_R = Dropout(0.5)(conv4_1_R)


##spatial Attentiin Right
c1=32
c1=int(drop4_2_L.shape[3]/8)
c=4
c=drop4_2_L.shape[2]
d=256
d=drop4_2_L.shape[3]
fa_2=Conv2D(c1,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_R)
fb_2=Conv2D(c1,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_R)
fc_2=Conv2D(d,1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(drop4_2_R)

fa1_2=tf.reshape(fa_2,[-1,c*c,c1])
fb1_2=tf.reshape(fb_2,[-1,c*c,c1])
fb1_2=tf.transpose(fb1_2, perm=[0, 2, 1])
fs_2=tf.linalg.matmul(fa1_2,fb1_2)
fs1_2=tf.nn.softmax(fs_2)
fc1_2=tf.reshape(fc_2,[-1,c*c,d])
t2=tf.linalg.matmul(fs1_2,fc1_2)
t3=tf.reshape(t2,[-1,c,c,d])
#Fsa_2=0.01*t3+drop4_2_R
#define eta right
e_r = tf.Variable(0.)
Fsa_2=e_r*t3+drop4_2_R

#channel attention Right
fch_R=tf.reshape(drop4_2_R,[-1,c*c,d])
fch1_R=tf.transpose(fch_R, perm=[0, 2, 1])
F_R=tf.linalg.matmul(fch_R,fch1_R)
F_R=tf.nn.softmax(F_R)
s_R=tf.linalg.matmul(fch1_R,F_R)
s_R=tf.reshape(s_R,[-1,c,c,d])
#Fsc_2=0.01*s_R+drop4_2_R
#define gama left
g_r = tf.Variable(0.)
Fsc_2=g_r*s_R+drop4_2_R

#channel and spatial attention fusion in Right in parallel form

Z_Right=Fsc_2+drop4_2_R
Z_Right=Conv2D(512,3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(Z_Right)
ZZ1=Fsa_2+Z_Right
ZZ1=Conv2D(512,3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(ZZ1)
R_Right=drop4_2_R+ZZ1
Fs_merge_2 = R_Right


##creat distance map between Fsa1 and Fsa2 by euclidean distance distance map between Fsa1 and Fsa2 by euclidean distance
dist=tf.math.squared_difference(Fs_merge_1 ,Fs_merge_2)

# downsampling L,R
z1=concatenate([conv3_1_R ,conv3_1_L], axis=3)
up6 = Conv2d_BN(UpSampling2D(size=(2, 2))(dist), 512, 2)
merge6 = concatenate([z1,up6], axis = 3)
conv6 = Conv2d_BN(merge6, 512, 3)
conv6 = Conv2d_BN(conv6, 512, 3)
z2=concatenate([conv2_1_R ,conv2_1_L], axis=3)
up7 = Conv2d_BN(UpSampling2D(size=(2, 2))(conv6), 256, 2)
merge7 = concatenate([z2, up7], axis=3)
conv7 = Conv2d_BN(merge7, 256, 3)
conv7 = Conv2d_BN(conv7, 256, 3)
z3=concatenate([conv1_1_R ,conv1_1_L], axis=3)
up8 = Conv2d_BN(UpSampling2D(size=(2, 2))(conv7), 128, 2)
merge8 = concatenate([z3, up8], axis=3)
conv8 = Conv2d_BN(merge8, 128, 3)
conv8 = Conv2d_BN(conv8, 128, 3)
up9 = Conv2d_BN(UpSampling2D(size=(2, 2))(conv8), 128,2)
conv9 = Conv2d_BN(up9, 64, 3)
conv9 = Conv2d_BN(conv9, 64, 3)

conv10 = Conv2d_BN(conv9, classes, 1, use_activation=None)
activation = Activation('sigmoid', name='Classification')(conv10)
model = Model(inputs=input, outputs=activation)
opt = tf.keras.optimizers.Adam(learning_rate=1e-4)

model.summary()
```

```python
def crossentropy_Edge_Loss(y, dist):
    loss_1=K.binary_crossentropy(y, dist)

#loss_Edge

    edge_pre_1=MaxPooling2D(pool_size=(3, 3), strides=(1, 1), padding='same')(dist)
    edge_dist_1=edge_pre_1[:,:,:,1]-dist[:,:,:,1]
    edge_pre_0=MaxPooling2D(pool_size=(3, 3), strides=(1, 1), padding='same')(dist)
    edge_dist_0=dist[:,:,:,0]-edge_pre_0[:,:,:,0]

    #tensor_0=[dist[:,:,:,0],edge_pre_0[:,:,:,0]]
    #edge_dist_0=tf.norm(tensor_0, ord=1, axis=None, keepdims=None, name=None)


    edge_gt_1=MaxPooling2D(pool_size=(3, 3), strides=(1, 1), padding='same')(y)
    edge_gt_1=edge_gt_1[:,:,:,1]-y[:,:,:,1]
    edge_gt_0=MaxPooling2D(pool_size=(3, 3), strides=(1, 1), padding='same')(y)
    edge_gt_0=y[:,:,:,0]-edge_gt_0[:,:,:,0]

    #tensor_1=[y[:,:,:,0],edge_gt_0[:,:,:,0]]
    #edge_gt_0=tf.norm(tensor_1, ord=1, axis=None, keepdims=None, name=None)

    edge_dist=tf.stack((edge_dist_0, edge_dist_1),axis=-1)
    edge_gt=tf.stack((edge_gt_0, edge_gt_1),axis=-1)

    loss_2 = K.binary_crossentropy(edge_gt,edge_dist)

    loss = loss_1 +0.5*loss_2
    #teta = tf.Variable(0.)
    #loss = loss_1 +teta*loss_2


    return loss
```

```python
model.compile(optimizer = opt, loss = crossentropy_Edge_Loss, metrics = ['accuracy'])
print("model loaded")
```

```python
NO_OF_EPOCHS = 17
hist=model.fit(train_gen, epochs=NO_OF_EPOCHS,
                          steps_per_epoch = STEP_SIZE_TRAIN,
                          validation_data=test_gen,
                          validation_steps=STEP_SIZE_VAL)
```
